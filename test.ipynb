Check ln(scaled risk score) distribution

```python
ddr = pd.read_csv('validation data/DDR_full_net_risk_fixed.csv')

ddr['scaled_risk_score_log'] = np.log(ddr.loc[:,'scaled_risk_score'].to_numpy(dtype=float))

ddr
```

```python
import matplotlib.pyplot as plt

thr = round(np.percentile(ddr['scaled_risk_score_log'], 50),3)

_ = plt.hist(ddr['scaled_risk_score_log'], bins='auto', range=(-200,0),  color ='green')
plt.title('Distribution of scaled_risk_score_log of Drugs')
plt.xlabel('scaled_risk_score_log')
plt.ylabel('Fraction of Drugs')

plt.axvline(thr, 0, 0.35, color = 'white', ls = '--')
plt.text(thr+0.2,20,thr,fontdict=dict(color='white', size=10))

plt.show()
```

```python

_ = plt.hist(ddr['scaled_risk_score_log'], bins='auto', range=(-200,0),  color ='purple')
plt.title('Distribution of scaled_risk_score_log of Drugs on Log Scale')
plt.xlabel('scaled_risk_score_log')
plt.ylabel('log(Fraction of Drugs)')

plt.axvline(thr, 0, 0.8, color = 'white', ls = '--')
plt.text(thr+0.2,13,thr,fontdict=dict(color='white', size=10))

plt.yscale('log')

plt.show()
```
